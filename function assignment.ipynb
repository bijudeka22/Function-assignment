#1. What is the difference between a function and a method in Python?
- In Python, functions and methods are both blocks of code that perform specific tasks, but they differ in their association with objects:
- Functions are standalone blocks of code that can be called from anywhere in your program. They don't belong to any specific object.
- Methods are functions that are associated with a particular class or object. They are defined within a class and are called using dot notation (e.g., object.method()).
-Functions are independent, while methods are tied to objects.
-Methods have access to the object's internal data (attributes) through the self parameter.
-Functions are called by their name, while methods are called using dot notation on an object

```python
def func(x):  #function example
   return x * x
a=func(5) #func is a standalone function that can be called directly with an argument.
print(a)
```

```python
class Mclass: # Method example
    def __init__(self,x):
        self.x = x
    def my_method(self):
        return self.x * self.x
obj = Mclass (5)
result = obj.my_method() # my_method is a method defined within the MClass class.
print(result)            #It can only be called on an instance of the class (the obj object in this case).
```

#3 What are the different ways to define and call a function in Python?

- Required Arguments: When defining a function, if a parameter is not assigned a default value, it becomes a required argument.These arguments must be provided in the exact order they are defined within the function's parentheses when the function is called.
-  Default Arguments: These arguments have default values specified in the function definition. If no value is provided for a default argument when the function is called, the default value is used.

- Keyword Arguments: These arguments are passed by their name, which allows them to be provided in any order.

- Arbitrary Arguments: These arguments allow a variable number of arguments to be passed to a function. They are prefixed with an asterisk ( * ) for positional arguments or two asterisks ( ** ) for keyword arguments.

```python
def greet(name): # Since the name is not a given default value, it is a required arguiment
    print("Hellow,", name + "!")
greet("Biju") # Calling the function with required arguiment
```

#5.What are iterators in Python and how do they differ from iterables?

- iterator is an object that allows you to traverse a collection of data (like a list, tuple, or string) sequentially. It provides a way to access elements one by one without needing to load the entire collection into memory at once.
- Iterables are like a container of items. They provide the data you want to loop over.
-Iterators are like a pointer that moves through the container, giving you access to one item at a time.
- Every iterator is also an iterable, but not every iterable is an iterator.
-Iterators are essential for efficient memory usage and processing of large datasets.

```python

my_list = [1, 2, 3]  # my_list is an iterable

my_iterator = iter(my_list)  # Get an iterator from my_list

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# Calling next() again will raise a StopIteration exception
```

#6.Explain the concept of generators in Python and how they are defined?


- Generators are a special type of iterable object in Python. They don't store all the values in memory at once like lists or tuples. Instead, they generate values on-the-fly as needed.
- yield Keyword: The core mechanism of a generator is the yield keyword. When a function contains yield, it automatically becomes a generator.
- Memory Efficiency: Generators are highly memory-efficient because they produce values one at a time. This is crucial when dealing with large datasets.
- Laziness: Generators are "lazy" in that they only compute the next value when requested, making them suitable for infinite sequences.
- Concise Code: Generators can often lead to more concise and readable code compared to traditional approaches.

Generators are a powerful tool in Python for creating efficient and memory-friendly iterables. They are particularly useful when you need to deal with large datasets or infinite sequences.

```python

def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci(10):
    print(num)
```

#7. What are the advantages of using generators over regular functions?

-  Generators provide a powerful and flexible mechanism for creating efficient and memory-friendly iterables in Python. They are particularly useful when dealing with large datasets, infinite sequences, or situations where you need to control the generation of values on-the-fly. Generators produce values one at a time, as needed. This means they don't need to store all the values in memory at once, making them ideal for handling large datasets or infinite sequences. This can significantly reduce memory usage and prevent potential memory errors.

```python

def even_numbers(limit):
   num = 0
   while num < limit:
         yield num
         num += 2

for even in even_numbers(10):
    print(even)
```

 #9. Explain the purpose and usage of the `map()` function in Python.


-
The map() function in Python applies a given function to each item in an iterable (like a list, tuple, or set) and returns an iterator that yields the results.

- map(function, iterable)

 * function: This is the function that you want to apply to each element of the iterable. It can be any callable object, including a regular function or a lambda function.
 * iterable: This is the sequence of elements that the function will be applied to.

- The map() function in Python is a valuable tool for efficiently and elegantly applying transformations to collections of data. Its conciseness, efficiency, and readability make it a preferred choice for many programming tasks.

```python
numbers = [1,2,3,4,5]
squared_numbers = list(map(lambda x: x*x, numbers))
print(squared_numbers)
```

#10. What is the difference between `map()`, `reduce()`, and `filter()` fun
ctions in Python?

- 1. map()
 * Purpose: Applies a given function to each item in an iterable and returns an iterator containing the results.
 * Action: Transforms each element.

- 2. filter()
 * Purpose: Filters elements from an iterable based on a given condition (a function that returns True or False).
 * Action: Selects elements that meet a specific criterion.


- 3. reduce()
 * Purpose: Applies a function cumulatively to the items of an iterable, reducing them to a single value.
 * Action: Aggregates elements into a single result.

```python

numbers = [1, 2, 3, 4] # Map example
squared = map(lambda x: x * x, numbers)
list(squared)
```

```python

numbers = [1, 2, 3, 4, 5] #filer example
even_numbers = filter(lambda x: x % 2 == 0, numbers)
list(even_numbers)
```

```python

from functools import reduce # reduce example
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

#4. Write a Python function that checks if a given number is prime or not from 1 to 200.?

```python

def is_prime(n):

  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

for num in range(1, 201):
  if is_prime(num):
    print(num, "is prime")
```

#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python

class FibonacciIterator:


    def __init__(self, n):


        self.n = n
        self.count = 0
        self.prev = 0
        self.curr = 1

    def __iter__(self):

        return self

    def __next__(self):

        if self.count >= self.n:
            raise StopIteration

        self.count += 1
        result = self.prev
        self.prev, self.curr = self.curr, self.prev + self.curr
        return result


fib_iter = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for num in fib_iter:
    print(num)
```

#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python

def powers_of_two(exponent):

  power = 1
  for i in range(exponent + 1):
    yield power
    power *= 2

for power in powers_of_two(5):
  print(power)
```

#7.Implement a generator function that reads a file line by line and yields each line as a string.

```python

def read_lines_from_string(string):
    for line in string.splitlines():
        yield line

# Example usage:
my_string = "file\nThis\nis\na\nmultiline\nstring."
for line in read_lines_from_string(my_string):
  print(line)
```

#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.?

```python

def is_not_vowel(char):
  vowels = "aeiouAEIOU"
  return char not in vowels

my_string = "Hello, world!"
no_vowels = "".join(filter(is_not_vowel, my_string))

print(no_vowels)
```
